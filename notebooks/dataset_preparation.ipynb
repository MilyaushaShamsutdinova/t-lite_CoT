## Dataset loading

```python
from datasets import load_dataset

ds = load_dataset("Egor-AI/Russian_thinking_dataset")
```

```python
ds
```

```python
ds = ds["train"].train_test_split(test_size=0.2)
print(ds)
```

```python
ds["train"][0]
```

```python
def preprocess_dataset(example):
    messages = [
        {"role": "system", "context": example["system"]},
        {"role": "user", "context": example["prompt"]},
        {"role": "assistant", "context": example["response"]},
    ]
    return {"messages": messages}
```

```python
ds_train_processed = ds["train"].map(preprocess_dataset)
ds_test_processed = ds["test"].map(preprocess_dataset)
```

```python
ds_train_processed = ds_train_processed.remove_columns(["prompt", "response", "system"])
ds_test_processed = ds_test_processed.remove_columns(["prompt", "response", "system"])
```
